关于最后一层网络（输出层）：

- 任何类型网络最后一层 `一定` 是全连接层（tf.keras.layers.Dense）
- 二分类问题输出维度一定是 `1`
- 多分类问题输出维度是 `类别总数`
- 回归问题输出维度一定是 `1`
    1. 回归到任意值： `不带任何激活函数`
    2. 回归到0~1范围内的值： 激活函数一定是sigmoid
    
sigmoid: 将任意值压缩到 0 到 1 之前

- 什么是One-Hot？ 
> 比如总共有3个分类， 类别一：用数组[1, 0, 0]表示, 类别二：用数组[0, 1, 0]表示，类别三：用数组[0, 0, 1]表示。卷积神经网络就是采用这种编码来表示单标签的

- 什么是K-Hot？ 
> 比如总共有3个分类， 样本一属于类别1和类别2，那么用数组[1, 1, 0]表示

## 6. 补充

### 6.1 MSE

> MSE: Mean Squared Error  
均方误差是指参数估计值与参数真值之差平方的期望值;  
MSE可以评价数据的变化程度，MSE的值越小，说明预测模型描述实验数据具有更好的精确度。

$$MSE = \frac{1}{N}\sum_{i=1} ^N (observed_{i} - predictedt_{i} )^2 $$

### 6.2 RMSE

>RMSE  
均方根误差是均方误差的算术平方根

$$RMSE = \sqrt{MSE} =\sqrt {\frac {1} {N} \sum_{i=1}^N (observed_{i} - predictedt_{i} )^2} $$

### 6.3 MAE

> MAE :Mean Absolute Error  
平均绝对误差是绝对误差的平均值  
平均绝对误差能更好地反映预测值误差的实际情况.

$$MAE = \frac {1} {N} \sum_{i=1}^N |observed_{i} - predictedt_{i}|$$

### 6.4 SD

> SD :standard Deviation  
标准差:标准差是方差的算术平方根。标准差能反映一个数据集的离散程度。平均数相同的两组组数据，标准差未必相同。

$$SD = \sqrt{ \frac {1} {N} \sum_{i=1}^N (x_{i}- \overline{x})^2 }$$

$$ \overline{x} 表示平均值( mean )： \overline{x} = \frac {1} {N} (x_{1} + x_{2} + ..... + x_{N})$$
